```python
import pandas as pd

# Load the data
file_path = '/Users/judithyemeli/Documents/CSE_6242/Project/MVR/network_graph/inputs_data/high_sim.csv'
data = pd.read_csv(file_path)

# Create unique nodes
nodes = pd.DataFrame(pd.concat([data['id_1'], data['id_2']]).unique(), columns=['id'])
nodes['group'] = 'county'  # Assign a group label (for styling in D3)

# Create links with similarity scores
links = data[['id_1', 'id_2', 'similarity']].rename(columns={'id_1': 'source', 'id_2': 'target'})

# Convert to JSON-like structure for D3
graph_data = {
    "nodes": nodes.to_dict(orient='records'),
    "links": links.to_dict(orient='records')
}

# Output JSON structure for D3
graph_data
```
